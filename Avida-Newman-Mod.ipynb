```python
import pandas as pd
from parse_trace import tracefile_iter
import networkx as nx
```

```python
from networkx import graphviz_layout
```

```python
with open('test.trace') as fp:
    df = pd.DataFrame([row for row in tracefile_iter(fp)])
df.set_index('idx', inplace=True)
```

```python
gG = nx.Graph()

last_used_by = {'AX':None, 'BX':None, 'CX':None, 'S0':[], 'S1':[]}
last_val     = {'AX':None, 'BX':None, 'CX':None, 'S0':[], 'S1':[]}

for idx, row in df.iterrows():
    for reg in ['AX', 'BX', 'CX']:
        if row[reg] != last_val[reg]:
            if last_val[reg] is not None:
                first = last_used_by[reg]
                middle = idx
                last = row.IP
                gG.add_node(first, node_type='IP')
                gG.add_node(middle, node_type='exec', inst=row.inst, register=reg)
                gG.add_node(last, node_type='IP')
                gG.add_edge(first, middle, val=hex(last_val[reg]), len=10)
                gG.add_edge(middle, last, val=hex(row[reg]), len=10)
            last_used_by[reg] = row.IP
            last_val[reg] = row[reg]
```

```python
try:
    figsize(18,12)
    #pos = nx.spring_layout(gG, dim=2)
    gG.graph['overlap'] = 'scale'
    gG.graph['splines'] = 'true'
    pos = nx.graphviz_layout(gG, prog='neato')
    nx.draw_networkx_edges(gG, pos, alpha=0.4, width=1.5, edge_color='0.35')
    nx.draw_networkx_nodes(gG, pos, node_color='0.85', linewidths=2.0, node_size=400)
    
    labels = {}
    for node, data in gG.nodes_iter(data=True):
        if data['node_type'] is 'exec':
            labels[node] = data['inst'] + '\n' + data['register']
        else:
            labels[node] = 'IP:' + str(node)
    nx.draw_networkx_labels(gG, pos, font_size=10, labels=labels, )
    
    nx.draw_networkx_edge_labels(gG, pos, alpha=0.5, 
                             edge_labels={(n1, n2): dat['val'] for n1, n2, dat in gG.edges_iter(data=True)})
    print nx.to_agraph(gG)
except UnicodeWarning:
    pass
```
